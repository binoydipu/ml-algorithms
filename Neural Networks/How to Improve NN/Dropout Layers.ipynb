# Apply Dropout Layers in ANN

- Works when the neural network is overfitting

Dropout is a regularization technique in artificial neural networks where, during training, a fraction ùëù of neurons is randomly "dropped" (set to 0) in each forward pass. This prevents overfitting by reducing co-adaptation of neurons.

### Intuition

- Think of it as ‚Äúturning off‚Äù some neurons randomly while training.

- This forces the network to learn more robust, distributed representations instead of relying too much on specific neurons.

- At test time, all neurons are active, but their outputs are scaled by (1‚àíùëù) to maintain balance.

### Effect of ùëù:

- Small ùëù (e.g., 0.1‚Äì0.3): mild regularization, useful when dataset is small.

- Moderate ùëù (0.5): strong regularization, common in hidden layers.

- Large ùëù (>0.7): too many neurons dropped ‚Üí underfitting, model struggles to learn.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from mlxtend.plotting import plot_decision_regions
```

# Regression Problem

```python
X_train = np.linspace(-1, 1, 20)
```

```python
y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  , 
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475, 
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,  
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
```

```python
X_test = np.linspace(-1, 1, 20)
```

```python
y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,  
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
plt.scatter(X_train, y_train, c='red', label='Train')
plt.scatter(X_test, y_test, c='blue', label='Test')
plt.legend()
plt.show()
```

# Regression Model

```python
model_1 = Sequential([
    Dense(128, input_dim=1, activation="relu"),
    Dense(128, activation="relu"),
    Dense(1, activation="linear")
])

model_1.compile(loss='mse', optimizer=Adam(learning_rate=0.01), metrics=['mse'])
history = model_1.fit(
    X_train, y_train, 
    epochs=500, validation_data = (X_test, y_test), verbose=False)
```

```python
# evaluate the model
_, train_mse = model_1.evaluate(X_train, y_train, verbose=0)
_, test_mse = model_1.evaluate(X_test, y_test, verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))
```

```python
y_pred_1 = model_1.predict(X_test)
```

```python
plt.figure()
plt.scatter(X_train, y_train, c='black', label='Train')
plt.scatter(X_test, y_test, c='red', label='Test')
plt.plot(X_test, y_pred_1)
plt.legend()
plt.ylim((-1.5, 1.5))
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Normal Model (train vs test costs)')
```

# Dropout Model

```python
model_2 = Sequential([
    Dense(128, input_dim=1, activation="relu"),
    Dropout(0.2),
    Dense(128, activation="relu"),
    Dropout(0.2),
    Dense(1, activation="linear")
])

model_2.compile(loss='mse', optimizer=Adam(learning_rate=0.01), metrics=['mse'])

drop_out_history = model_2.fit(X_train, y_train, epochs=500, 
                               validation_data = (X_test, y_test),
                               verbose=False)
```

```python
# evaluate the model
_, train_mse = model_2.evaluate(X_train, y_train, verbose=0)
_, test_mse = model_2.evaluate(X_test, y_test, verbose=0)
print('Train: {}, Test: {}'.format(train_mse, test_mse))
```

```python
y_pred_2 = model_2.predict(X_test)
```

```python
plt.figure()
plt.scatter(X_train, y_train, c='black', label='Train')
plt.scatter(X_test, y_test, c='red', label='Test')
plt.plot(X_test, y_pred_2)
plt.legend()
plt.ylim((-1.5, 1.5))
plt.show()
```

```python
plt.plot(drop_out_history.history['loss'])
plt.plot(drop_out_history.history['val_loss'])
plt.title('Dropout (20%) Model (train vs test costs)')
```
